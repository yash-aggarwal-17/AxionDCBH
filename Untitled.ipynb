## Define v=0 and LTE Collisional Dissociation 

This first part compares v=0 and LTE collisional dissociation rates of H2 molecule. 

```python
# Dissociation via electron impact 

# v=0 
def D_ele_0 (T): 
    # units: cm^3 s^(-1) 
    val = 4.49e-9 * T**(0.11) * np.exp(-101858 / T)
    return val


# LTE 
def D_ele_LTE (T): 
    # units: cm^3 s^(-1) 
    val = 1.91e-9 * T**(0.136) * np.exp(-53407.1 / T)
    return val 
```

```python
# Dissociation via HI impact 

# v = 0 
def D_HI_0 (T): 
    # units: cm^3 s^(-1)  
    val = 6.67e-12 * T**(0.5) * np.exp(- (1 + 63593/T))
    return val 

# LTE
def D_HI_LTE (T):
    # units: cm^3 s^(-1)  
    val = 3.52e-9 * np.exp( -43900 / T)
    return val 
```

## H2 Formation Collisional rates

```python
# Define H+e- -> H- rate 

def R_Hm (T): 
    # units cm^3 s^(-1)
    T4 = T/1e+4
    
    val = 1.65e-18 * T4 ** (0.76 + 0.15*np.log(T4) - 0.033 * (np.log(T4))**2 )
    return val
    
```

## Collisional Heating and Cooling rates 

```python
# Define Low Density Cooling Rate per H2 molecule 
def lambda_low_n(T):
    """ 
    Returns cooling rate per H2 molecule in low density regime. 
    units: ergs cm^3 s^(-1)
    """
        
    x=np.log10(T)
    val = 10**( - 103.0 + 97.59*x - 48.05*(x)**2 + 10.80 * (x)**3 - 0.9032*(x)**4 )
    return val 
```

```python
lambda_low_n(3000)
```

```python
# Plot dissociation rates 
temp_list = 10**np.linspace(2,4,50)

plt.figure(figsize=(7.2,6.0))
plt.loglog()

# Electron dissociation rates 
plt.plot(temp_list, D_ele_0(temp_list), label=r'e$^{-1}$ ($\nu$=0)', linestyle='dashed')
plt.plot(temp_list, D_ele_LTE(temp_list), label=r'e$^{-1}$ (LTE)')

# HI dissociation rates 
plt.plot(temp_list, D_HI_0(temp_list), label=r'HI ($\nu$=0)', linestyle='dashed')
plt.plot(temp_list, D_HI_LTE(temp_list), label=r'HI (LTE)')

# H- rate 
plt.plot(temp_list, R_Hm(temp_list), label=r'H$^{-}$')

# H2 cooling rate 
plt.plot(temp_list, lambda_low_n(temp_list))

plt.legend()

plt.ylim(1e-20,1e-8)
plt.xlim(1e+2, 1e+4)

plt.ylabel(r'Collisional rates (cm$^3$ s$^{-1}$)')
plt.xlabel(r'Temperature (K)')

plt.show()
```
